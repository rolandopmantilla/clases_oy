## An치lisis de tabla de contingencia

```python
Peliculas['Budget_bin']=pd.qcut(Peliculas["Budget (million $)"],q=4)
```

```python
pd.crosstab(index=Peliculas['Budget_bin'], columns='N칰mero')
```

```python
tbl_cont=pd.crosstab(Peliculas['Budget_bin'], Peliculas['Genre'])
tbl_cont
```

```python
chi_test=scipy.stats.chi2_contingency(tbl_cont)
chi_test
```
